```python

def process_histogramm(image):
    # Convertir la imagen a escala de grises
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Procesar el histograma de la imagen en escala de grises
    processed_gray_image = cv2.equalizeHist(gray_image)

    # Crear una imagen en color con el histograma procesado
    processed_image = cv2.cvtColor(processed_gray_image, cv2.COLOR_GRAY2BGR)
    
    return processed_image

def process_histogram(image):
    channels = cv2.split(image)
    processed_channels = []
    for channel in channels:
        processed_channel = cv2.equalizeHist(channel)
        processed_channels.append(processed_channel)
    processed_image = cv2.merge(processed_channels)
    
    return processed_image

for i in range(1,5):
  
    imageColor = cv2.imread(f'./images/{i}.png', cv2.IMREAD_COLOR)#cv2.IMREAD_COLOR

    imageGray = cv2.imread(f'./images/{i}.png', cv2.COLOR_BGR2GRAY)

    processed_histogram_Gray = process_histogramm(imageGray)

    processed_histogram_Color = process_histogram(imageColor)

    plt.figure(figsize=(8, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Imagen Original')

    plt.subplot(1, 2, 2)
    plt.imshow(processed_histogram_Color, cmap='gray')#processed_histogram
    plt.title('Histograma Color')


    plt.show()

    plt.figure(figsize=(8, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Imagen Original')

    plt.subplot(1, 2, 2)
    plt.imshow(processed_histogram_Gray, cmap='gray')#processed_histogram
    plt.title('Histograma a Escala Grises')


    plt.show()
```

```python

def arithmetic_operators(image1, image2,operator):

    if operator == 1:
        result_image = cv2.add(image1, image2)

    elif operator == 2:
        result_image = cv2.subtract(image1, image2)

    elif operator == 3:
        result_image = cv2.multiply(image1, image2)

    elif operator == 4:
        result_image = cv2.divide(image1, image2)
    return result_image


for i in range(1,5):
    image1 = cv2.imread(f'./images/{i}.png', cv2.IMREAD_GRAYSCALE)
    image2 = cv2.imread(f'./images/{i}.png', cv2.IMREAD_GRAYSCALE)

    for j in range(1,5):
        if j == 1:
            OpeArit = 'Suma'
        elif j == 2:
            OpeArit = 'Resta'
        elif j == 3:
            OpeArit = 'Multiplicacion'
        elif j == 4:
            OpeArit = 'Division'
        arithmetic_result = arithmetic_operators(image1, image2,j) # Por ejemplo, se puede cambiar el operador aquí
 
        plt.figure(figsize=(12, 4))

        plt.subplot(1, 3, 1)
        plt.imshow(image1, cmap='gray')
        plt.title('Primera Imagen')

        plt.subplot(1, 3, 2)
        plt.imshow(image2, cmap='gray')
        plt.title('Segunda Imagen')

        plt.subplot(1, 3, 3)
        plt.imshow(arithmetic_result, cmap='gray')
        plt.title(f'Op. Aritmética Tipo:{OpeArit} Imagen:{i}')

        plt.show()
```

```python

def arithmetic_operators_color(image1, image2, operator):


    if operator == 1:
        result_image = cv2.add(image1, image2)
    elif operator == 2:
        result_image = cv2.subtract(image1, image2)
    elif operator == 3:
        result_image = cv2.multiply(image1, image2)
    elif operator == 4:
        result_image = cv2.divide(image1, image2)
                
    return result_image    

for i in range(1,5):

    image1Color = cv2.imread(f'./images/{i}.png', cv2.IMREAD_COLOR)
    image2Color = cv2.imread(f'./images/{i}.png', cv2.IMREAD_COLOR)
    for j in range(1,5):
        if j == 1:
            OpeArit = 'Suma'
        elif j == 2:
            OpeArit = 'Resta'
        elif j == 3:
            OpeArit = 'Multiplicacion'
        elif j == 4:
            OpeArit = 'Division'
        arithmetic_result_color = arithmetic_operators_color(image1Color, image2Color, j)
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.imshow(cv2.cvtColor(image1Color, cv2.COLOR_BGR2RGB))
        plt.title('Primera Imagen')

        plt.subplot(1, 3, 2)
        plt.imshow(cv2.cvtColor(image2Color, cv2.COLOR_BGR2RGB))
        plt.title('Segunda Imagen')

        plt.subplot(1, 3, 3)
        plt.imshow(cv2.cvtColor(arithmetic_result_color, cv2.COLOR_BGR2RGB))
        plt.title(f'Op. Aritmética Tipo:{OpeArit} Imagen:{i}')

        plt.show()
```
